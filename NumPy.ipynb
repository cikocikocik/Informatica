```python
# ============================================================
# INTRODUZIONE A NUMPY: CREAZIONE, ANALISI E MANIPOLAZIONE DI ARRAY
# ============================================================

import numpy as np

# ------------------------------------------------------------
# 1) CREAZIONE DI UN ARRAY NUMPY 1D (VETTORE)
# ------------------------------------------------------------
a = np.array([1, 2, 3])

print(a)              # Stampa dell'array
print(a[1])           # Accesso all'elemento all’indice 1 (secondo elemento)
print(type(a))        # Tipo della variabile (numpy.ndarray)
print(type(a[1]))     # Tipo dell'elemento interno (numpy.int64 o simile)
print(a.dtype)        # Tipo dei dati contenuti nell'array
```

```python
# ------------------------------------------------------------
# 2) CREAZIONE DI UN ARRAY 2D (MATRICE)
# ------------------------------------------------------------
a2 = np.array([
    [1, 2, 3],
    [5, 6, 7],
    [7, 8, 9],
    [10, 11, 12]
])

print(a2)
```

```python

# ------------------------------------------------------------
# 3) PROPRIETÀ DEGLI ARRAY
# ------------------------------------------------------------

# Numero di dimensioni (1 = vettore, 2 = matrice)
print("Dimensionalità dell'array:")
print(a.ndim)     # 1 dimensione
print(a2.ndim)    # 2 dimensioni

# Forma (righe, colonne)
print("Shape:")
print(a.shape)    # (3,)
print(a2.shape)   # (4, 3)

# Numero totale di elementi
print("Size:")
print(a.size)     # 3 elementi
print(a2.size)    # 12 elementi (4x3)
```

```python
# ------------------------------------------------------------
# 4) ACCESSO AGLI ELEMENTI
# ------------------------------------------------------------
print(a2[1])      # Seconda riga della matrice
print(a2[1][0])   # Primo elemento della seconda riga
```

```python
# ------------------------------------------------------------
# 5) ARRAY 2D DI UNA SOLA RIGA
# ------------------------------------------------------------
a3 = np.array([[1, 2, 3]])
print(a3)
print(a3.shape)   # (1, 3) → 1 riga, 3 colonne
```

```python

# ------------------------------------------------------------
# 6) ARRAY DI ZERI
# ------------------------------------------------------------
z = np.zeros((3, 4))   # Matrice 3x4 riempita di 0
print(z)
```

```python




# ------------------------------------------------------------
# 7) ARRAY CON RANGE
# np.arange(start, stop, step)
# ------------------------------------------------------------
r = np.arange(1, 10, 2)
print(r)
```

```python

# ------------------------------------------------------------
# 8) RESHAPING DI UN ARRAY
# ------------------------------------------------------------
a = np.array([1, 2, 3, 4])
print(a)
print(a.reshape(2, 2))   # Da vettore 1D a matrice 2x2
```

```python

# ------------------------------------------------------------
# 9) RAVEL: "appiattisce" un array multidimensionale
# ------------------------------------------------------------
print(a2.ravel())   # Ritorna un array 1D con tutti gli elementi
print(a2)           # L'array originale NON viene modificato
```

```python

# ------------------------------------------------------------
# 10) ITERAZIONE SUGLI ELEMENTI
# ------------------------------------------------------------

# Iterazione su TUTTI gli elementi tramite flatten()
for e in a2.flatten():
    print(e)

# Iterazione solo sulla seconda riga
for e in a2[1]:
    print(e)
```

```python

# ------------------------------------------------------------
# 11) SOMMA DEGLI ELEMENTI
# ------------------------------------------------------------
print(a2.sum())       # Somma totale

# Somma sulle righe (axis = 1)
print("Somma per righe:")
print(a2.sum(axis=1))
```
