```python
# Variabili semplici
num = 10
print(num)
print(type(num))

num1 = 3.9
print(type(num1))

stringa = "ciao"
print(type(stringa))

# Liste
l = [1, 2, 3, 4]
print(type(l))

# Creazione lista vuota e aggiunta elementi
l2 = list()
l2.append("c")
l2.append(1)
print(l2)

# Aggiunta di un'intera lista dentro una lista
l2.append(l)
print(l2)

l2 = 2
print(l2)
print(type(l2))
# Convertire una stringa in lista di caratteri
l3 = list("ciao")
print(l3)
l7 = ["ciao"]
print(l7)

c = 'r'
print(type(c))

d = "r" 
print(type(d))

ciao = l3[0] + l3[1] + l3[2] + l3[3]
print(ciao)

# Creazione liste con list()
l4 = list([1, 2, 3])
print(l4)

l5 = list(["ciao"])
print(l5)

# Lunghezza della lista e index()
print(len(l3))
print(l3.index("a"))

# Inserimento fuori range → aggiunge in coda
l3.insert(6, 'Z')
print(l3)

# pop() senza indice rimuove l’ultimo elemento
l3.pop()
print(l3)

# Questo genera un errore perché l'indice 7 non esiste
l3.pop(7)  # IndexError: pop index out of range
```
